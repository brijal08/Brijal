## Q1 - For the below data build a Naive Bayes classifier and identify the class of the given data sets noting down the probabilities as the model would save them.

```python
class NaiveBayesClassifier:
    def __init__(self):
        self.class_probabilities = {}
        self.feature_probabilities = {}

    def fit(self, X_train, y_train):
        # Calculate class probabilities
        total_samples = len(y_train)
        classes, class_counts = np.unique(y_train, return_counts=True)
        for class_, class_count in zip(classes, class_counts):
            self.class_probabilities[class_] = class_count / total_samples

        # Calculate feature probabilities for each class
        for feature_index in range(X_train.shape[1]):
            self.feature_probabilities[feature_index] = {}
            unique_values = np.unique(X_train[:, feature_index])
            for class_, class_count in zip(classes, class_counts):
                self.feature_probabilities[feature_index][class_] = {}
                for value in unique_values:
                    # Count occurrences of value in class
                    class_mask = y_train == class_
                    value_count = np.sum(X_train[class_mask][:, feature_index] == value)
                    # Calculate probability of feature value given class
                    self.feature_probabilities[feature_index][class_][value] = value_count / class_count

    def predict(self, X_test):
        predictions = []
        for sample in X_test:
            max_probability = -1
            predicted_class = None
            for class_, class_prob in self.class_probabilities.items():
                probability = class_prob
                for feature_index, feature_value in enumerate(sample):
                    if feature_index in self.feature_probabilities:
                        if class_ in self.feature_probabilities[feature_index] and feature_value in self.feature_probabilities[feature_index][class_]:
                            probability *= self.feature_probabilities[feature_index][class_][feature_value]
                        else:
                            # If feature value not seen in training, set probability to zero
                            probability = 0
                            break
                if probability > max_probability:
                    max_probability = probability
                    predicted_class = class_
            predictions.append(predicted_class)
        return predictions
```

## Q2 - For the below data what would be the prediction for play using a Naive Bayes classifier?Please use Gaussian PDF for the continuous variables


#### Create Gaussian naive bayes classifier class

```python
class GaussianClf:
    def __init__(self): # Intialization function
        self.lables = None
        self.priors = {}
        self.means = {}
        self.variances = {}
        
    def fit(self,X,y): #train model
        self.lables = np.unique(y)
        for lable in self.lables:
            X_label = X[y == lable]
            self.priors[lable] = len(X_label)/len(X)
            self.means[lable] = np.mean(X_label,axis=0)
            self.variances[lable] = np.var(X_label,axis=0)
        print('Lables:',self.lables)
        print('Means:',self.means)
        print('Varience:',self.variances)
            
    def calculateProbability(self,x,mean,variance): # calculate probability
        probability = 0
        try:
            exponent = -0.5 * ((x-mean)**2) / variance
            probability = (1.0/np.sqrt(2 * np.pi * variance)) * np.exp(exponent)
        except ValueError:
            print(ValueError, " Probability:",probability)
        return probability
    
    def predict(self,X): #Predict model for output
        predictions = []
        for x in X:
            classScore = {}
            for label in self.lables:
                prior = np.log(self.priors[label])
                mean = self.means[label]
                variance = self.variances[label]
                
                probability = self.calculateProbability(x,mean,variance)
                classScore[label] = np.sum(np.log(probability)) + prior
            
            predictedLabel = max(classScore,key=classScore.get)
            predictions.append(predictedLabel)
        
        return predictions
```

### Create and train the Bayes Optimal classifier

```python
# Function to calculate the prior probabilities of each class
def calculate_prior_probabilities(df):
    class_counts = df['Classification'].value_counts()
    total_samples = df.shape[0]
    priors = {}
    for class_label, count in class_counts.items():
        priors[class_label] = count / total_samples
    return priors

# Function to calculate the likelihood probabilities of each feature for each class
def calculate_likelihood_probabilities(df, feature, class_label):
    feature_counts = df[df['Classification'] == class_label][feature].value_counts()
    total_samples_in_class = df[df['Classification'] == class_label].shape[0]
    likelihoods = {}
    for feature_value in df[feature].unique():
        if feature_value in feature_counts:
            likelihoods[feature_value] = feature_counts[feature_value] / total_samples_in_class
        else:
            likelihoods[feature_value] = 0
    return likelihoods

# Function to make predictions for a new instance
def predict_instance(instance, priors, likelihoods):
    best_class = None
    max_posterior_prob = -1
    
    for class_label in priors:
        posterior_prob = priors[class_label]
        
        for feature in instance:
            posterior_prob *= likelihoods[feature][class_label][instance[feature]]
            
        if posterior_prob > max_posterior_prob:
            max_posterior_prob = posterior_prob
            best_class = class_label
    
    return best_class
```

```python

# Calculate prior probabilities
priors = calculate_prior_probabilities(df)

# Calculate likelihood probabilities
likelihoods = {}
for feature in X.columns:
    likelihoods[feature] = {}
    for class_label in priors:
        likelihoods[feature][class_label] = calculate_likelihood_probabilities(df, feature, class_label)
```

```python
#### Predict the values using model
Test = [ 
    {'running nose': 0, 'coughing': 1, 'reddened skin': 0, 'fever': 1},
    {'running nose': 1, 'coughing': 0, 'reddened skin': 0, 'fever': 1},
    {'running nose': 1, 'coughing': 0, 'reddened skin': 1, 'fever': 0}
]
Test
```

```python
predictions = []
for x in Test:
    predicted_class = predict_instance(x, priors, likelihoods)
    predictions.append(predicted_class)
print("Predicted Classes:", predictions)
```
